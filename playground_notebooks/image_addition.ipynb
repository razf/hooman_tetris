```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_im = cv2.imread(r"1.png")
cutout = cv2.imread(r"sample_cutout.png", cv2.IMREAD_UNCHANGED)
base_im = cv2.cvtColor(base_im, cv2.COLOR_BGR2RGB)
#base_im[cutout[:,:,3] != 0] = 255
plt.figure(figsize=(10,10))
plt.imshow(base_im)
```

```python
contours, hierarchy = cv2.findContours(cutout[:,:,3], cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours_im = cv2.drawContours(base_im, contours, -1, (255,255,255), 10)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(contours_im)
```
